```python
# sigma is a RBF spread parameter and simga > 0
def Gaussian(X, sigma):
    return np.exp(-(np.linalg.norm(X,axis = -1)/sigma)**2)
```

```python
def RBF(W, args):
    sigma = args[1]
    X = args[2]
    C = W[0]
    v = W[1]
    #print(v.shape, Gaussian(X-C,sigma).shape)
    return np.matmul(v,Gaussian(X-C, sigma).T) 
```

```python
def loss(W, args):
    N = args[0]
    v = W[-N:]
    a = W[:-N].reshape(1,N,2)
    y = args[3]
    c = np.repeat(a,args[2].shape[0],axis=1).reshape((args[2].shape[0],N,2))
    

    regu = np.sum(a**2) + np.sum(v**2)
    print(regu)
    return float((1/(2*N)* np.sum((RBF((c,v), args) - y.T)**2, axis=1) + args[4]/2* regu)[0])
```

```python
def preproc(N, X_train):
    a =  np.random.normal(size=(1, N, 2))
    res = (np.tile(X_train,(N,1)).reshape((X_train.shape[0],N,2)))
    
    #C = np.repeat(a,X_train.shape[0], axis=1).reshape(X_train.shape[0],N,2)
    

    v = np.random.normal(size = N)
    return a,v,res 
```

```python
def test_error(W,  args):
    N = args[0]
    v = W[-N:]
    a = W[:-N].reshape(1,N,2)
    y = args[3]
    c = np.repeat(a,args[2].shape[0],axis=1).reshape((args[2].shape[0],N,2))

    return float((1/(2*N)* np.sum((RBF((c,v), args) - y.T)**2, axis=1)))
```

```python
## main loop
rho = 1e-6
sigma = 5.5
N = 5
C,V,res = preproc(N, X_train)

shape = C.shape
args = (N,sigma, res, y_tr, rho, shape)
W = np.append(C.flatten(),V)

print(loss(W, args))
result_rbf = scipy.optimize.minimize(loss, W, args= (args,) , method='CG')#, options = {"maxiter": 5})
```

```python
print(result_rbf)
```

```python
###testing settings:
args = (N, sigma, res, y_tr, rho, shape)
ter = test_error(result_rbf.x, args)
print(ter)
```

# Optimization in Python


Suppose we want to find the global minimum (Does it exist? Is it unique? Why? Why not?) of this function:

$$4x_{1}^2+4x_{1}x_{2}+10x_{2}^2+12x_{1}+4x_{2}$$ 

It is a quadratic strictly convex function. It has one and only one global minimum.<br> Here if you want the proof https://math.stackexchange.com/questions/337090/if-f-is-strictly-convex-in-a-convex-set-show-it-has-no-more-than-1-minimum

We define the function and plot it 

```python
def fun_plot(X,Y): 
    #X and Y are matrices
    return 4 * X ** 2 + 4 * X * Y + 10 * Y ** 2 + 12 * X + 4 * Y

```

```python
plotting(fun_plot, 'Optimize it')
```

If we want to minimize the function, we have to write it so that it takes in input a single variable.

```python
def fun(X): 
    x1 = X[0]
    x2 = X[1]
    return 4 * x1 ** 2 + 4 * x1 * x2 + 10 * x2 ** 2 + 12 * x1 + 4 * x2


def fun_grad(X):
    x1 = X[0]
    x2 = X[1]

    dx1 = 8*x1 + 4*x2 + 12
    dx2 = 4*x1 + 20*x2 + 4
    return np.array([dx1,dx2])
```

Now we are interested in solving this simple problem with the easisest iterative method: __the gradient method with fixed stepsize!__

<div class="alert alert-block alert-success">
<b> Gradient method with fixed stepsize:</b> <br> Given a starting point $x_0, k=0$ and a threshold $\epsilon>0$<br>

while $\Vert \nabla f(x^k)\Vert>\epsilon$ 
$x^{k+1}=x^k-\alpha^k\nabla f(x^k)$.
</div>


```python
def gradient_method_fixed_alpha(fun, x_0,alpha=0.0001,max_it=1000,toll=1e-6):
    """
    - fun is the function to optimize
    - x_0 is the starting point
    - alpha is the fixed stepsize
    - max_it is the maximum number of iterations
    - toll is the tollerance on the optimization
    """
    opt_time = time.time()
    x = x_0
    k = 0
    flag = True
    while np.linalg.norm(fun_grad(x))>toll:
        x = x-alpha*fun_grad(x)
        k+=1
        if k > max_it:
            flag = False
            break
    
    print('Time spent to optimize the function:',time.time()-opt_time,'sec')
    print('Starting point:',x_0)
    print('Stationary point found:',flag)
    print('Final point found:',x)
    print('Objective value at the final point:', fun(x),'gradient:',np.linalg.norm(fun_grad(x)))
    return x, fun(x), iter, flag #final point found, number of iterations, flag to know if the stationary point has been found
```

Now we can try to solve the problem by applying the gradient method:

```python
x_0 = np.random.rand(2)
x, value, iter, flag = gradient_method_fixed_alpha(fun,x_0,0.1) 
```

How is it possible that the objective function is so high? And the gradient? Something went wrong for sure. Let's use a smaller stepsize!

_NB The problem is on the fact that, to guarantee convergence to a stationary point, $\alpha^k$ must be smaller than certain threshold_

```python
x, value, iter, flag = gradient_method_fixed_alpha(fun,x_0,0.01) 
```

## Scipy
We do not need too implement our own optimization function since we can rely on __scipy__

The SciPy library provides many user-friendly and efficient numerical routines, such as routines for numerical integration, interpolation, optimization, linear algebra, and statistics.

Let's introduce the __scipy.optimize__ package. 

SciPy optimize provides functions for minimizing (or maximizing) objective functions, possibly subject to constraints. It includes solvers for nonlinear problems (with support for both local and global optimization algorithms), linear programing, constrained and nonlinear least-squares, root finding, and curve fitting. You can find info here: https://docs.scipy.org/doc/scipy/reference/tutorial/optimize.html

There are several algorithms we can choose from to optimize our function.
For example, we can find the minimum with the __Nelder and Mead__ algorithm.

```python
x0 = np.array([5,8])   #define a starting point
```

```python
from scipy.optimize import minimize 

start = time.time()

res = minimize(fun,x0,method='nelder-mead')  #the results of the optimization are saved in the "res" variable. 
                                              # Print it to see what it looks like!

print("Time for optimizing:: %.5f" % (time.time()-start),"seconds")

print('Optimal Point:',res.x, '\nObjective function value:',fun(res.x),'\ngradient value:', np.linalg.norm(fun_grad(res.x)))

print(res)
```

__Nelder and Mead__ is a _derivative-free_ algorithm and it is not the most suitable algorithm in our differentiable setting!
Since the problem is a quadratic strictly convex function we could apply a __Conjugate Gradient__ method which is a _state of the art_ algorithm for this kind of problems

```python
from scipy.optimize import minimize 

start = time.time()

res = minimize(fun,x0,method='CG',jac=fun_grad)

print("Time for optimizing:: %.5f" % (time.time()-start),"seconds")

print('Optimal Point:',res.x, '\nObjective function value:',fun(res.x),'\ngradient value:', np.linalg.norm(fun_grad(res.x)))
```

Notice how CG returns a better solution (both in terms of $f(x)$ and $\Vert \nabla f(x)\Vert$) in a smaller amount of time!


__Notice this: with Scipy, we can pass only one variable to the function we want to minimize. This is why we need to be smart when the function takes other parameters in input.__



We want to optimize the trivial function$$y =a_1x_1^2+a_2x_2^2$$
where $a_1\geq0$ and $a_2\geq0$ are two hyperparameters chosen by me <br>


Note: when defining the function we want to optimize, the first input __MUST BE__ the variable we want to optimize and it has to be a __vector__! (Of shape (n,))<br>
If the variable of the problem is composed by different blocks($X=X_1,X_2)$ and we need to have them separately in the objective function, we can wrap them into an array, pass it to the solver and then unwrap them again in the function.

```python
def fun(X, functionArgs):  # X is a vector
    a1 = functionArgs[0] # unwrapping the arguments
    a2 = functionArgs[1]
    x1 = X[0] # unwrapping the variable
    x2 = X[1]
    return a1*x1**2 + a2*x2**2
```

Define the starting point $X_0$ and the hyperparameters

```python
x0 = np.array([1,3])

a1 = 1
a2 = 2

functionArgs = [a1,a2]
```

The first element we have to pass to the $minimize$ function must be the function we want to optimize; the second has to be the starting point, while in $args$ I can put the hyperparameters, if there are some. NB: __Hyperparameters are something I choose!__

```python
import time

start=time.time()

minimize(fun, x0, args=functionArgs,method='bfgs')

running_time=time.time()-start

print(running_time)
```

$\begin{aligned}
& \underset{x}{\text{minimize}}

& & 4x_{1}-3x_{2}=1312\\
& & x_{1}+x_{2}\geq 100\\
& & x_{1}\geq 0\\
& & x_{2}\geq 0\\
\end{aligned}
$
